# 구별 인구수

```python
import pandas as pd # Python Data Analysis Library
```

```python
import json # JSON encoder and decoder
import folium # Python library used for visualizing geospatial data
```

```python
from folium import plugins 
```

```python
# import library: time
import time
now = time.localtime() # used to add timestamp to filename
```

```python
#folium.__version__
```

```python
gupop = pd.read_excel('./files/gupop.xlsx')
```

```python
gupop.head()
```

```python
gupop.sort_values(by='val', ascending=True).head(25)
```

```python
seoulmappath  = './files/seoul_sgg.geojson'
seoulmap = json.load(open(seoulmappath, encoding='utf-8'))
```

```python
loc = 'Populatiom by Distict, SEOUL'
loc = '서울 구별 인구수'
title_html = '''
             <head>
             <meta charset="utf-8">
             </head>
             <h3 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(loc) 
```

```python
facChoro = folium.Map(
    location = [37.58585751211208, 126.97624264491466],
    tiles = 'cartodbpositron',
    zoom_start=11
)

choropleth = folium.Choropleth(
    geo_data = seoulmap,
    data = gupop,
    columns=['dist', 'val'],
    fill_color = 'YlOrBr',
    fill_opacity=0.7,
    line_opacity=0.5,
    key_on='properties.SIG_KOR_NM'
    ).add_to(facChoro)


choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(['SIG_KOR_NM'])
)

```

```python
facChoro.get_root().html.add_child(folium.Element(title_html))
```

```python
minimap = plugins.MiniMap()
facChoro.add_child(minimap)
```

```python
facChoro
```

```python
output_path = './output/'
file_name = 'vis-facChoro'
rel_path =  output_path + file_name + '_' + str(now.tm_mon).zfill(2) + str(now.tm_mday).zfill(2) + str(now.tm_hour).zfill(2) + str(now.tm_min).zfill(2) + '.html'
```

```python
facChoro.save(rel_path) 
```
